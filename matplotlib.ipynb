```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,6])
y = np.array([0,80])
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Charts
## Line Chart
points = np.array([2,3,4,6,7,8,2,4,5])
plt.plot(points)
```

```python
a_points = np.array([2,3,4,6,7,8,2,4,5])
plt.plot(a_points, linestyle='dotted')
```

```python
# Bar chart
x = np.array(['A','B','C','D'])
y = np.array([1,2,3,4])


plt.bar(x,y)
```

```python
# Scattered plot

x1 = np.array([2,4,6,8])
y1 = np.array([1,3,5,7])
plt.scatter(x1,y1)
```

```python
# PIE CHART
p = np.array([30,20,40,10])

plt.pie(p)
```
